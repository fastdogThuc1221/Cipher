# Encryption

```
import random
alphabet = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ .!?1234567890'
def randomize_alphabet(alphabet):
  key = list(alphabet)
  random.shuffle(key)
  return ''.join(key)

print(randomize_alphabet(alphabet))
```

```
plain_text = "A black hole is a region of spacetime exhibiting gravitational acceleration so strong that nothing—no particles or even electromagnetic radiation such as light—can escape from it.[6] The theory of general relativity predicts that a sufficiently compact mass can deform spacetime to form a black hole.[7][8] The boundary of the region from which no escape is possible is called the event horizon. Although the event horizon has an enormous effect on the fate and circumstances of an object crossing it, no locally detectable features appear to be observed.[9] In many ways, a black hole acts like an ideal black body, as it reflects no light.[10][11] Moreover, quantum field theory in curved spacetime predicts that event horizons emit Hawking radiation, with the same spectrum as a black body of a temperature inversely proportional to its mass. This temperature is on the order of billionths of a kelvin for black holes of stellar mass, making it essentially impossible to observe."
print(plain_text)
```

```
def build_dictionary(left, right):
  dictionary = {}
  for i in range(len(left)):
    dictionary[left[i]] = right[i]
  return dictionary

key = randomize_alphabet(alphabet)
print(build_dictionary(alphabet, key))
```

```

def encrypt_subsitution_cipher(alphabet, key, plain_text):
  dictionary = build_dictionary(alphabet, key)
  cipher_text = ''
  for i in range(len(plain_text)):
    letter = plain_text[i]
    if letter in dictionary:
      cipher_text = cipher_text + dictionary[letter]
    else:
      cipher_text = cipher_text + letter
  return cipher_text

cipher_text = encrypt_subsitution_cipher(alphabet, key, plain_text)
print(cipher_text)
```

# Decryption

```
def decrypt_substitution_cipher(alphabet, key, cipher_text):
  return encrypt_subsitution_cipher(key, alphabet, cipher_text)
print(decrypt_substitution_cipher(alphabet, key, cipher_text))
```
