Chapter 4 from Cracking Codes with Python

```
def reverse_cipher(plain_text):
  n = len(plain_text)
  cipher_text = ''
  for i in range(n):
    cipher_text = cipher_text + plain_text[n-1-i]
  return cipher_text
```

```
# Testing
s = 'xin chao ban Long. Hom nay la ngay dep troi lam. Thuc dang hoc Python.'
print(reverse_cipher(s))
```

```
t = reverse_cipher(s)
print(reverse_cipher(t))
```
