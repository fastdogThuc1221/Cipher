```
def shift_letter(input_letter, offset):
  index = ALPHABET.find(input_letter)
  new_index = index + offset
  if new_index > len(ALPHABET):
    new_index = new_index - len(ALPHABET)
  if new_index < 0:
    new_index = new_index + len(ALPHABET)
  #print(new_index)
  new_letter = ALPHABET[new_index]
  return new_letter

print ('Letter:  {}. Shifted letter: {}'.format('a', shift_letter('a', 3)))
print ('Letter:  {}. Shifted letter: {}'.format('z', shift_letter('z', 20)))
```

```
def shift_message(message, offset):
  output_message = ''
  for letter in message:
    output_message = output_message + shift_letter(letter, offset)
  return output_message
```

```
cipher_text = shift_message("MY Secret message", 3)
text = shift_message(cipher_text, -3)
print(cipher_text)
```
