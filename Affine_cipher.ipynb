# Modular Addition and Subtraction and Mulitiplication

##(A + B) mod C = (A mod C + B mod C) mod C


```
# A = 14, B = 17, C = 5
(14 + 17) % 5
```

```
((14 % 5) + (17 % 5)) % 5
```

## (A - B) mod C = (A mod C - B mod C)

```
# A = 14, B = 17, C = 5
(14 - 17) % 5
```

```
((14 % 5) - (17 % 5)) % 5
```

## (A x B) mod C = (A mod C x B mod C) 


```
# A = 14, B = 17, C = 5
(14 * 17) % 5
```

```
((14 % 5) * (17 % 5)) % 5
```

# Modular Inverse

>$A * A^{-1}  mod  C = 1$

```
def modular_inverse(key, divisor):
  for x in range(divisor):
    result = (key * x) % divisor
    if result == 1:
      return x
  print('Modular inverse not found.')
  return None
```

```
modular_inverse(2, 6)
modular_inverse(3, 7)
```

#Encryption

```
ALPHABET = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890 !?.'
plain_text = 'I am very happy!!'
def modulo_letter(key, letter):
  index =  ALPHABET.find(letter)
  new_index = (index*key) % len(ALPHABET)
  new_letter = ALPHABET[new_index]
  return new_letter
```

```
def encrypt_affine(key, plain_text):
  cipher_text = ''
  for letter in plain_text:
    cipher_text = cipher_text + modulo_letter(key, letter)
  return cipher_text
```

```
cipher_text = encrypt_affine(13, plain_text)
print(cipher_text)
```

# Decryption

```
def decrypt_affine(key, cipher_text):
  inverse_key = modular_inverse(key, len(ALPHABET))
  plain_text = ''
  for letter in cipher_text:
    plain_text = plain_text + modulo_letter(inverse_key, letter)
  return plain_text
```

```
modular_inverse(13, len(ALPHABET))
```

```
decrypt_affine(13, cipher_text)
```

```
plain_text = 'In mathematics, in particular the area of number theory, a modular multiplicative inverse of an integer a is an integer x such that the product ax is congruent to 1 with respect to the modulus m.'
cipher_text = encrypt_affine(13, plain_text)
decrypted_text = decrypt_affine(13, cipher_text)
print(cipher_text)
print(decrypted_text)
```
